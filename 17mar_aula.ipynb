```python
# operadores aritiméticos

# pedir ao usuário para digitar 02 números


#entrada

a = int (input("digite um número: "))
b = int (input("digite um número: "))


# processamento

# soma

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b


# saída

print("")
print("------------OPERADORES ARITIMÉTICOS------------------------------")
print("")
print("---------------SOMA----------------------------------------------")
print("")
print(f"O resultado da SOMA de {a} MAIS {b} é igual a : {soma}")
print("")
print("---------------------SUBTRAÇÃO------------------------------------")
print("")
print(f"O resultado da DIFERENÇA de {a} MENOS {b} é igual a : {subtracao}")
print("")
print("---------------------MULTIPLICAÇÃO--------------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} VEZES {b} é igual a : {multiplicacao}")
print("")
print("---------------------DIVISÃO------------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO {b} é igual a : {divisao}")
print("")
print("---------------------DIVISÃO------------------------------------")
print("")
print(f"O resultado da DIVISÃO INTEIRA de {a} DIVIDIDO POR {b} é igual a : {divisao}")
print("")
print("---------------------RESTO------------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO {b} o RESTO é igual a : {resto}")
print("")
print("-----------------------------------------------------------------------")
print("")
print("---------------------POTÊNCIA------------------------------------")
print("")
print(f"O resultado da POTENCIAÇÃO de {a} ELEVADO {b} é igual a : {potencia}")
print("")
print("-----------------------------------------------------------------------")
```
